```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import time
import math
```

# Question 2a)

```python
def load_data():
    df_real = pd.read_csv('clean_real.txt', names =["Text"])
    df_real['Real?']=1 
    df_fake = pd.read_csv('clean_fake.txt', names =["Text"])
    df_fake['Real?']=0
    df = pd.concat([df_real,df_fake])
    
    
    from sklearn.feature_extraction.text import CountVectorizer

    # Initialize the "CountVectorizer" object
    vectorizer = CountVectorizer()

    # fit_transform() does two functions: First, it fits the model
    # and learns the vocabulary; second, it transforms our training data
    # into feature vectors. 

    train_data_features = vectorizer.fit_transform(df.Text)

    # Numpy arrays are easy to work with, so convert the result to an 
    # array
    train_data_features = train_data_features.toarray()
    # Get the names of all the words 
    vocab = vectorizer.get_feature_names()
    
    # Create a dataframe called the train_test_data
    train_test_data=pd.DataFrame(data=train_data_features,columns=vocab)
    
    #Assign dataframe to X and Whether fake or real news to y
    X = train_test_data
    y = df['Real?']
    # Add y to X Dataframe called Real?
    X['Real?'] = y.tolist()
    # Randomly shuffle the rows in X dataframe
    X=X.sample(frac=1)
    
    # Take 70%, 15% and 15& of rows 
    df1 = X.iloc[:int(len(X)*0.7)]
    df2 = X.iloc[int(len(X)*0.7):int(len(X)*0.85)]
    df3 = X.iloc[int(len(X)*0.85):]
                 
    # Assign the split dataframe to training, validating and testing datasets
    # called X_train,y_train,X_val,y_val,X_test and y_test
    y_train = df1['Real?']
    X_train = df1.drop(['Real?'],axis = 1) 
                 
    y_val = df2['Real?']
    X_val = df2.drop(['Real?'],axis = 1)
                 
    y_test = df3['Real?']
    X_test = df3.drop(['Real?'],axis = 1) 
    
    return X_train,y_train,X_val,y_val,X_test,y_test,vocab
```

```python
def compute_information_gain(Xi):
    X_train,y_train,X_val,y_val,X_test,y_test,vocab = load_data()
    # Add the fake or real column to the training set
    X_train['Real?']=y_train.tolist()
    
    # Number of total samples 
    total = len(X_train)
    
    # How many fake and real news in sample
    rt_fake_count = len(X_train.loc[X_train['Real?'] == 0])
    rt_real_count = len(X_train.loc[X_train['Real?'] == 1])
    
    # Calculate entropy of root node
    En_root = -rt_fake_count/total * math.log2(rt_fake_count/total) -rt_real_count/total * math.log2(rt_real_count/total)
    
    # For the given number of column, find all the ones that
    # contain the word and all the ones that does not contain the word, 
    # this would be our left child sample and right child sample
    X_contain = X_train[X_train[Xi]  == 1]
    X_not_contain = X_train[X_train[Xi]  == 0]
    
    # Calculate left child fake and real news title number and total sample
    left_fake_count = len(X_contain.loc[X_contain['Real?'] == 0])
    left_real_count = len(X_contain.loc[X_contain['Real?'] == 1])
    left_total = len(X_contain)
    
    # Compute Entropy of Left Child
    left_root = -left_fake_count/left_total * math.log2(left_fake_count/left_total)-left_real_count/left_total * math.log2(left_real_count/left_total)
    
    
    # Calculate right child fake and real news title number 
    right_fake_count = len(X_not_contain.loc[X_not_contain['Real?'] == 0])
    right_real_count = len(X_not_contain.loc[X_not_contain['Real?'] == 1])
    right_total = len(X_not_contain)
    
    # Compute Entropy of Right Child
    right_root = -right_fake_count/right_total * math.log2(right_fake_count/right_total) -right_real_count/right_total * math.log2(right_real_count/right_total)
    
    # Compute the information gain of the split
    IG = En_root - (len(X_contain)/total*left_root + len(X_not_contain)/total*right_root)
    
    print('Informaition Gain is about '+ str(IG) + ' with the word split ' + str(Xi))
```

```python
compute_information_gain('donald')
compute_information_gain('trumps')
compute_information_gain('the')
compute_information_gain('america')
compute_information_gain('clinton')
```
